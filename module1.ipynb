```python
high_pop=df.sort_values(by='Population' ,ascending=False).head(10)
high_pop
```

```python
df['literacy_rate']=(df['Literate']/df['Population'])*100
lit_rate=df[['State_name','District_name','Population','literacy_rate']]
lit_rate=lit_rate.sort_values(by='literacy_rate',ascending=False)
lit_rate.head(10)
```

```python
df['gender_ratio'] = (df['Female'] / df['Male']) * 1000
gen_ratio=df[['State_name','District_name','Female','Male' ,'gender_ratio']]
gen_ratio.sort_values(by="gender_ratio",ascending=False)
```
