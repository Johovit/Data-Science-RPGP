```python
df['team_size'] =df['team_size'].fillna(df['team_size'].median())
df['login_frequency_per_month']=df['login_frequency_per_month'].fillna(df['login_frequency_per_month'].median())
df['customer_satisfaction_score']=df['customer_satisfaction_score'].fillna(df['customer_satisfaction_score'].median())

df['company_name']=df['company_name'].fillna(df['company_name'].mode()[0])
df['last_support_ticket_date']= df['last_support_ticket_date'].fillna('No Ticket')
df['support_ticket_category']=df['support_ticket_category'].fillna('Unknown')

df['churn_date']=df['churn_date'].fillna('No Churn')
```

```python
df.isnull().sum()
```

```python
subscription_plan = df['subscription_plan']
```

```python
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for col in df.columns:
    if df[col].dtype=='object':
        df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
scaler = MinMaxScaler()
scaled_df=scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca_result =pca.fit_transform(scaled_df)

df_pca=pd.DataFrame(pca_result,columns=['PC1','PC2'])

print("Explained variance:",pca.explained_variance_ratio_.sum()*100)
```

```python
df_pca['subscription_plan']=subscription_plan

for i in df_pca['subscription_plan'].unique():
    subset=df_pca[df_pca['subscription_plan']==i]
    plt.scatter(subset['PC1'],subset['PC2'],label=str(i))

plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.2f})')
plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.2f})')
plt.title('PCA of saas Dataset')
plt.legend()
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda =LinearDiscriminantAnalysis(n_components=2)
lda_result=lda.fit_transform(scaled_df,subscription_plan)

df_lda= pd.DataFrame(lda_result,columns=['LD1', 'LD2'])

print("Explained variance:",lda.explained_variance_ratio_.sum()*100 )
```

```python
df_lda['subscription_plan']=subscription_plan.values

for i in df_lda['subscription_plan'].unique():
    subset =df_lda[df_lda['subscription_plan']==i]
    plt.scatter(subset['LD1'],subset['LD2'] ,label=str(i))

plt.xlabel(f'Linear Discriminant 1({lda.explained_variance_ratio_[0]*100:.2f})')
plt.ylabel(f'Linear Discriminant 2({lda.explained_variance_ratio_[1]*100:.2f})')
plt.title('LDA of saas Dataset')
plt.legend()
plt.show()
```

**SUMMARY**

*   The histogram of monthly revenue shows a highly right-skewed distribution, with most customers contributing relatively low revenue, while a few outliers contribute significantly higher amounts.

*   The violin plot of feature adoption rates by subscription plan reveals that adoption patterns are fairly similar across plans, though the Enterprise plan shows slightly higher median adoption compared to others.

*   Missing values were addressed using median imputation for numeric fields and mode or placeholder values for categorical data.

*   Features were encoded and scaled to prepare them for dimensionality reduction.

*   From the PCA and LDA plots, we can see that most subscription plans have similar customer patterns, as their points overlap a lot, meaning they share many common behaviors. However, Enterprise customers stand out more in the LDA plot, showing that they have different usage, revenue, or engagement compared to others. The other plans (Premium, Pro, Basic, Starter) look more alike, making them harder to separate. This suggests Enterprise customers could be targeted with specific strategies, while other plans may need more detailed features to tell them apart.
