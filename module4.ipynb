```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.info()
```

```python
mean_age=int(df['Age'].mean().round())
df['Age']=df['Age'].fillna(mean_age)

df['Embarked']=df['Embarked'].fillna(method='ffill')

df=df.drop('Cabin',axis=1)
```

```python
df.isnull().sum()
```

```python
df
```

```python
original_df=df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
```

```python
le=LabelEncoder()
df['Embarked']=le.fit_transform(df['Embarked'])
df['Sex']=le.fit_transform(df['Sex'])
```

```python
df
```

```python
df=original_df
```

```python
ohe = OneHotEncoder()
```

```python
sex_ohe =ohe.fit_transform(df[['Sex']]).toarray()
sex_ohe
```

```python
sex_ohe_df =pd.DataFrame(sex_ohe,columns=ohe.get_feature_names_out())
sex_ohe_df
```

```python
df=pd.concat([df,sex_ohe_df],axis=1)
```

```python
df
```

```python
df=original_df
```

```python
oe1 = OrdinalEncoder(categories=[['Q','S','C']])
em_oe=oe1.fit_transform(df[['Embarked']])
em_oe_df=pd.DataFrame(em_oe,columns=['Embarked_Encoded'])
df=pd.concat([df,em_oe_df],axis=1)
df=df.drop('Embarked',axis=1)
df
```

```python
oe2 = OrdinalEncoder(categories=[['male','female']])
sex_oe=oe2.fit_transform(df[['Sex']])
sex_oe_df=pd.DataFrame(sex_oe,columns=['Sex_Encoded'])
df=pd.concat([df,sex_oe_df],axis=1)
df=df.drop('Sex',axis=1)
df
```
