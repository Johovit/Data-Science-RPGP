```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('India Census 2011.csv')
```

```python
df1.head()
```

```python
high_pop_df=df1.sort_values(by='Population' ,ascending=False)
high_pop_df=high_pop_df[['State_name','District_name','Population']]
top_pop=high_pop_df.head(10)
top_pop
```

```python
sns.barplot(data=top_pop, x='Population', y='District_name',color='#9E3FCC')
plt.title('Top 10 Districts by Population')
plt.xlabel('Population')
plt.ylabel('District')
plt.show()
```

```python
df1['literacy_rate']=(df1['Literate']/df1['Population'])*100
lit_rate=df1[['State_name','District_name','Population','literacy_rate']]
lit_rate=lit_rate.sort_values(by='literacy_rate',ascending=False)
top_lit=lit_rate.head(10)
top_lit
```

```python
sns.barplot(data=top_lit, x='literacy_rate', y='District_name',color='#7F8D27')
plt.title('Top 10 Districts by Literacy Rate')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('District')
plt.show()
```

```python
df1['gender_ratio'] = (df1['Female'] / df1['Male']) * 1000
gen_ratio=df1[['State_name','District_name','Female','Male' ,'gender_ratio']]
top5_gen=gen_ratio.sort_values(by="gender_ratio",ascending=False).head()
last5_gen=gen_ratio.sort_values(by="gender_ratio",ascending=False).tail()
```

```python
top5_gen
```

```python
last5_gen
```

```python
plt.suptitle('Top and Bottom 5 Districts by Gender Ratio(num of females per 1000 males)')

plt.subplot(2, 1, 1)
sns.barplot(data=top5_gen, x='gender_ratio', y='District_name',color='#237B37')
plt.title('Top 5 Districts')

plt.subplot(2, 1, 2)
sns.barplot(data=last5_gen, x='gender_ratio', y='District_name',color='#7B2323')
plt.title('Bottom 5 Districts')

plt.subplots_adjust(hspace=0.5)
plt.show()
```
